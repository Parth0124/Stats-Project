## Visualizing the cleaned_demographics.csv

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('../Stats Dataset Cleaned/cleaned_demographics.csv')
```

```python
df.head()
```

```python
### PLotting the number of assylums in every country.

sns.pairplot(df.drop(columns=['Year', 'Territory of asylum or residence', 'Location Name']), corner=True)
plt.show()
```

```python
### dropping the non numeric features

df.drop(columns=['Territory of asylum or residence', 'Location Name'], axis=1, inplace=True)
```

```python
df.corr()
```

```python
### plotting a heatmap with the realtion amongst the features.

corr_matrix = df.corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Demographic Data')
plt.show()
```
